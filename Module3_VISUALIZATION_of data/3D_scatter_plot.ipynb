```python
%matplotlib notebook
#%matplotlib gtk

import matplotlib.pyplot as plt
import pandas as pd
import matplotlib
from mpl_toolkits.mplot3d import Axes3D
df=pd.read_excel('cement.xls')
df
df.columns=['cement','slag','ash','water','superplasticizer','co_aggr','fin_aggr','age','strength']
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.set_xlabel('cement')
ax.set_ylabel('water')
ax.set_zlabel('strength')
plt.suptitle('cement x water x strength')
ax.scatter(df.cement,df.water,df.strength,c='g',marker='.')
```
