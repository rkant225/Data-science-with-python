```python
# .. your code here ..
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from mpl_toolkits.mplot3d import Axes3D
plt.style.use('ggplot')
df=pd.read_csv('/home/rahul/Desktop/data_science/EDX_Data_Science/jupyter/Module3_VISUALIZATION_of data/Datasets/wheat.data')
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')


ax.set_xlabel('area')
ax.set_ylabel('perimeter')
ax.set_zlabel('asymmetry')
plt.suptitle('area x perimeter x asymmetry')
ax.scatter(df.area,df.perimeter,df.asymmetry,c='r',marker='o')

ax.set_xlabel('width')
ax.set_ylabel('groove')
ax.set_zlabel('length')
plt.suptitle('width x groove x length')
ax.scatter(df.width,df.groove,df.length,c='g',marker='o')

```
