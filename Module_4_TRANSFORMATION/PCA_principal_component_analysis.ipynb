```python
import pandas as pd
import numpy as np
df=pd.read_csv('kideny.csv')
df
df=pd.DataFrame([df.age,df.bp,df.sg,df.pcv,df.wc])
df.fillna(0,inplace=True)
#Subtract mean to each sample to normalize the data
M=(df-df.T.mean(axis=1)).T
#Find the covatiance mattrix row x row
C=np.cov(M)

ev,coff=np.linalg.eig(C)
```
