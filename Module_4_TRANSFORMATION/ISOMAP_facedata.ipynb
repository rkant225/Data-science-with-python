```python
import math

import pandas as pd
import scipy.io
from sklearn import manifold
#Load the mattrix
mat = scipy.io.loadmat('Datasets/face_data.mat')
#convert to DataFrame
df = pd.DataFrame(mat['images']).T
#Find no. of images,no. of pixel,no. of pixels_per_dimension
num_images, num_pixels = df.shape
pixels_per_dimension = int(math.sqrt(num_pixels))

# Rotate the pictures
for idx in df.index:
    df.loc[idx] = df.loc[idx].values.reshape(pixels_per_dimension, pixels_per_dimension).T.reshape(-1)
    
df

```

```python


iso = manifold.Isomap(n_neighbors=10, n_components=2)
iso.fit(df)
manifold_temp = iso.transform(df)
T_isomap = pd.DataFrame(manifold_temp, columns=['Component1', 'Component2'])

# Left with 2 dimensions
T_isomap.shape
```

```python
from sklearn.decomposition import PCA
df
pca = PCA(n_components=2,svd_solver='full')
pca.fit(df)
T_pca=pca.transform(df)
# Left with 2 dimensions
T_pca = pd.DataFrame(T_pca, columns=['Component1', 'Component2'])

T_pca.shape
```

```python
import matplotlib.pyplot as plt
import matplotlib


T_pca.plot.scatter(x='Component1', y='Component2', marker='o', c='r', alpha=0.75)
T_isomap.plot.scatter(x='Component1', y='Component2', marker='o', c='g', alpha=0.75)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

fig = plt.figure()
fig.set_size_inches(10, 10)
ax = fig.add_subplot(111)
ax.set_title('2D Components from Isomap of Facial Images')
ax.set_xlabel('Component: 1')
ax.set_ylabel('Component: 2')

# Show 40 of the images ont the plot
x_size = (max(T_isomap['Component1']) - min(T_isomap['Component1'])) * 0.08
y_size = (max(T_isomap['Component2']) - min(T_isomap['Component2'])) * 0.08
for i in range(40):
    img_num = np.random.randint(0, num_images)
    x0 = T_isomap.loc[img_num, 'Component1'] - (x_size / 2.)
    y0 = T_isomap.loc[img_num, 'Component2'] - (y_size / 2.)
    x1 = T_isomap.loc[img_num, 'Component1'] + (x_size / 2.)
    y1 = T_isomap.loc[img_num, 'Component2'] + (y_size / 2.)
    img = df.iloc[img_num,:].values.reshape(pixels_per_dimension, pixels_per_dimension)
    ax.imshow(img, aspect='auto', cmap=plt.cm.gray, 
              interpolation='nearest', zorder=100000, extent=(x0, x1, y0, y1))

# Show 2D components plot
ax.scatter(T_isomap['Component1'], T_isomap['Component2'], marker='.',alpha=0.7)

ax.set_ylabel('Up-Down Pose')
ax.set_xlabel('Right-Left Pose')

plt.show()
```
